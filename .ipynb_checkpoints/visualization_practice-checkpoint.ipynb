Pretty Distribution Comparison below works, but I ended up preferring the outcome of the ridgeline plot. Also, you can just pass in the dataframe to ridgeline, so its better.

```python
def pretty_distribution_comparison(distributions_list, labels_list):
    
    sns.set(style='darkgrid')
    
    for idx, distribution in enumerate(distributions_list):
        fig = sns.kdeplot(distribution, shade = True)
        
    plt.legend(labels = labels_list)
    plt.xlabel('Math Score')
    plt.ylabel('Score Density')
    plt.title('Math Score Distributions by Tested Groups')
    plt.show()
```

```python
math_A = data.math_score[data.group == 'group A']
math_B = data.math_score[data.group == 'group B']
math_C = data.math_score[data.group == 'group C']
math_D = data.math_score[data.group == 'group D']
math_E = data.math_score[data.group == 'group E']
math_list = [ math_A, math_B, math_C, math_D, math_E]
math_labels = [ 'math_A', 'math_B', 'math_C', 'math_D', 'math_E']
```

```python
pretty_distribution_comparison(math_list, math_labels)
```

```python
def find_closest(num,arr):
    curr = arr[0]
    for index in range(len(arr)):
        if abs (num - arr[index]) < abs (num - curr):
            curr = arr[index]
    return curr
```

```python
def ridgeline( dataframe, score_col ):
    
    group_dict = { 0 : 'group A',
                   1 : 'group B',
                   2 : 'group C',
                   3 : 'group D',
                   4 : 'group E'
                 }
    
    for value in group_dict.values():
        group_mean = dataframe[dataframe['group'] == value][score_col].mean()
        dataframe.loc[dataframe['group'] == value,'group_mean_score'] = group_mean
        
    dataframe_sorted = dataframe.sort_values('group')
    
    pal = sns.color_palette( palette = 'Set3', n_colors = 5 )
    
    g = sns.FacetGrid( dataframe_sorted, row = 'group', hue = 'group_mean_score', 
                      aspect = 8, height = 2, palette = pal )
    
    g.map( sns.kdeplot, score_col, bw_adjust = .5, clip_on = False, fill = True, alpha = 1, 
          linewidth = 1.5, cut = 0 )
    g.map( sns.kdeplot, score_col, bw_adjust = .5, clip_on = False, color = 'w', lw = 2,
          cut = 0 )
    g.map( plt.axhline, y = 0, lw = 2, clip_on = False )
    
    for idx, ax in enumerate(g.axes.flat):
        ax.text(10,0.02,group_dict[idx],fontweight = 'bold', fontsize = 24, 
                color=ax.lines[-1].get_color())
        group_mean = dataframe_sorted.loc[dataframe_sorted['group'] == group_dict[idx],score_col].mean()
        adjusted_mean = find_closest(group_mean,ax.lines[0].get_data()[0])
        mean_index = np.where(ax.lines[0].get_data()[0] == adjusted_mean)
        h_at_mean_idx = ax.lines[0].get_data()[1][mean_index]
        normalized_h_at_mean = (h_at_mean_idx / ax.lines[0].get_data()[1].max()) * .5
        ax.axvline( x = group_mean, ymax = normalized_h_at_mean, c = 'black', clip_on = False,
                  label = group_mean)
        
    g.fig.subplots_adjust(hspace=-0.0)
    g.set_titles("")
    g.set(yticks=[])
    g.despine(bottom = True, left = True)
    
    
    plt.setp(ax.get_xticklabels(), fontsize=16, fontweight='bold')
    plt.xlabel('Student Test Score', fontweight='bold', fontsize=15)
    plt.xlim(0,100)
    subject = score_col.split('_')[0].capitalize()
    g.fig.suptitle(f'Test Score Probability Distributions by Testing Group in {subject}',
               ha='center',
               fontsize=20,
               fontweight=20)

    plt.show()
```

```python
ridgeline(data, 'math_score')
```
